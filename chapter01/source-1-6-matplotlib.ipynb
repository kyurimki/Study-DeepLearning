### 1.6.1 단순한 그래프 그리기
- matplotlib의 pyplot 모듈 이용

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0, 6, 0.1)  # 0~6까지 0.1 간격으로 생성
y = np.sin(x)

plt.plot(x, y)
plt.show()
```

### 1.6.2 pyplot의 기능

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0, 6, 0.1)
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1, label="sin")
plt.plot(x, y2, linestyle="--", label="cos")  # cos 함수는 점선으로 그리기
plt.xlabel("x")  # x축 이름
plt.ylabel("y")  # y축 이름
plt.title('sin & cos')  # 제목
plt.legend()
plt.show()
```
